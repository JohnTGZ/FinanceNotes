# Notes on present value

Plot present value of $100 at risk-free rate of 5% across a period of 0 to 10 years

```python
import numpy as np
import matplotlib.pyplot as plt

L = 100.0 # Principal

# R: Zero rate
R = 0.05 

t_i = np.arange(0, 5.0, 1.0) # 0 to 10 years
y = L * np.exp(-R * t_i)

plt.plot(t_i, y)

plt.show()
```
